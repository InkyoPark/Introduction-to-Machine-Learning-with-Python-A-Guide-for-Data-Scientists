```python
import mglearn
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.preprocessing import QuantileTransformer, StandardScaler, PowerTransformer
```

```python
#for example

mglearn.plots.plot_scaling()
```

```python
cancer = load_breast_cancer()

X_train, X_test, Y_train, Y_train = train_test_split(cancer.data, cancer.target, random_state = 1)

print(X_train.shape)
print(X_test.shape)
```

```python
scaler = MinMaxScaler()
scaler.fit(X_train)
```

```python
# Data Transform


X_train_scaled = scaler.transform(X_train)
# after Scaled print atribute
print("transformed voiume: \n", X_train_scaled.shape)
print("  : \n", X_train_scaled.shape)
print("\nbefore scaled min: \n", X_train.min(axis = 0))
print("\n    : \n", X_train.min(axis = 0))
print("\nbefore scaled max: ",X_train.max(axis = 0))
print("\n    : \n", X_train.max(axis = 0))
print("\nafter scale min: \n", X_train_scaled.min(axis = 0))
print("\n    : \n", X_train_scaled.min(axis = 0))
print("\nafter scale max: \n", X_train_scaled.max(axis = 0))
print("\n    : \n", X_train_scaled.max(axis = 0))
```

```python
# Test Data Transform


X_test_scaled = scaler.transform(X_test)
# after Scaled print atribute
print("\nafter scale min: \n", X_test_scaled.min(axis = 0))
print("\n    : \n", X_test_scaled.min(axis = 0))
print("\nafter scale max: \n", X_train_scaled.max(axis = 0))
print("\n    : \n", X_test_scaled.max(axis = 0))
```

```python
#Quantile Transformer and PowerTransformer


X, y = make_blobs(n_samples=50, centers=2, random_state=4, cluster_std=1)
X += 3
plt.scatter(X[:,0], X[:,1], c=y, s=30, edgecolors = "black")
plt.xlim(0,16)
plt.ylim(0,10)
plt.title("Original Data")
plt.show()
```

```python
scaler = QuantileTransformer()
X_trans = scaler.fit_transform(X)

plt.scatter(X_trans[:,0], X_trans[:,1], c=y, s=30, edgecolors = "black")
plt.xlim(0,5)
plt.ylim(0,5)
plt.title("Original Data -> QuantileTransformer")
plt.show()
```

```python
plt.hist(X_trans)
plt.show()
```

```python
print(scaler.quantiles_.shape)
```

```python
x = np.array([[0], [7], [8], [9], [10]])
print(np.percentile(x[:,0], [0, 25, 50, 75, 100]))
```

```python
X_trans = QuantileTransformer().fit_transform(x)
print(np.percentile(X_trans[:,0], [0, 25, 50, 75, 100]))
```

```python
# output_distribution = "uniform"  i think this is default
# output_distribution = "normal" 

scaler = QuantileTransformer(output_distribution = "normal") 
X_trans = scaler.fit_transform(X)

plt.scatter(X_trans[:,0], X_trans[:,1], c = y, s = 50, edgecolors = "black")
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.title(type(scaler))
plt.show()
```

```python
plt.hist(X)
plt.title("Original Data")
plt.show()

plt.hist(QuantileTransformer(output_distribution = "normal").fit_transform(X))
plt.title("QuantileTransformer")
plt.show()

plt.hist(StandardScaler().fit_transform(X))
plt.title("StandardScaler")
plt.show()

plt.hist(PowerTransformer(method= "box-cox").fit_transform(X))
plt.title("PowerTransformer box-cox")
plt.show()

plt.hist(PowerTransformer(method="yeo-johnson").fit_transform(X))
plt.title("PowerTransformer yeo-johnson")
plt.show()
```
